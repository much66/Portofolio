# 2. Hitunglah BMI setiap siswa dengan rumus : \begin{align}
        BMI = \frac{(ùêµùëíùëüùëéùë° ùëèùëéùëëùëéùëõ (ùëòùëî))}{(ùëáùëñùëõùëîùëîùëñ ùêµùëéùëëùëéùëõ (ùëö))^2}
    \end{align}

```python
berat = df_4['Berat Badan (kg)']
tinggi = df_4['Tinggi Badan (cm)']
```

```python
for i in range(len(berat)):
    TB = tinggi[:i+1]/100
    bmi = berat[:i+1] / (TB**2)
```

```python
bmi
```

# 5. Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI.
- Normal: 18.5-24.9
- Overweight: 25.0-29.9
- Obesity: 30.0-34.9
- Underweight: < 18.5


```python
status_list = []

for j in bmi:
    if j >= 18.5 and j <= 24.9:
        status_list.append("Normal")
    elif j >= 25.0 and j <= 29.9:
        status_list.append("Overweight")
    elif j >= 30.0 and j <= 34.9:
        status_list.append("Obesity")
    elif j < 18.5:
        status_list.append("Underweight")
```

```python
df_4.insert(loc=6, column='Status BMI', value=status_list)
```

# Program Investasi Jamin-Sugih Co

Perusahaan Investasi Jamin-Sugih Co kami memberikan penawaran investasi terbaik dengan pembagian keuntungan tetap sebesar 25% setiap tahun.

Buatlah program Python yang dapat menerima input dari pengguna dalam bentuk:
Modal Investasi (bilangan bulat antara 0-1.000.000.000)
Jangka Waktu Investasi (bilangan bulat antara 1-25 dalam tahun)
Selain itu, program akan menampilkan error message bahwa modal atau tenure tidak sesuai
ketentuan.

```python
Modal = float(input("Money Capital : Rp "))
Jangka_waktu = int(input("Investment Tenure (years) : "))
```

```python
if (Modal > 0 and Modal <= 1000000000) and (Jangka_waktu >= 1 and Jangka_waktu <= 25):
    modal_awal = Modal
    for i in range(Jangka_waktu):
        modal_awal = modal_awal * 125 / 100
        print("Year {:,.0f},".format(i+1),"your total investment Rp {:,.2f}".format(modal_awal))
    print("Nilai investasi anda setelah", Jangka_waktu, "tahun adalah Rp {:,.2f}".format(modal_awal))
else:
    print("Modal atau jangka waktu tidak sesuai dengan ketentuan")
```
